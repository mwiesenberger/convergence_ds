```python
import json
import simplesimdb as database
import yaml
import numpy as np
import pandas as pd
from IPython.display import display, HTML # to display tables in for loop

pd.set_option('display.float_format', lambda x: '%.2e' % x)
```

```python
#define conversion function 
def orderToString(x): 
    if np.isnan(x) : return 'n/a'
    return'%.2f'% x

def errorToString(x):
    return '%.2e' % x
```

```python
# Create Database manger
db = database.Manager( directory = 'data_ds_guenter_t',
        filetype='yaml', executable='./execute_guenter.sh')
```

```python
# First generate some data
support=[1,12]
methods=["dg", "linear", "linear-nearest", "linear-linear" ]
resolutionDG = [ (10,5), (16,10), (26,20), (40,40), (64,80) ]#, (100,160)]
resolutionL =  [ (8,5), (15,10), (30,20), (60,40), (120,80) ]#, (100,160)]
# create simulation database
for method in methods :
    for m in support :
        if m == 1  and method == "linear-nearest":
                continue
        if m == 1  and method == "linear-linear":
                continue
        if method == "dg" :
            for n in resolutionDG :
                inputfile = { "n": 3, "Nx" : n[0], "Ny" : n[0], "Nz" : n[1],
                    "mx" : m, "my" : m, "method" : method }
                print( inputfile)
                db.create( inputfile, error="display")
        else :
            for n in resolutionL :
                inputfile = { "n": 3, "Nx" : n[0], "Ny" : n[0], "Nz" : n[1],
                    "mx" : m, "my" : m, "method" : method }
                print( inputfile)
                db.create( inputfile, error="display")
                    
```

```python
content = db.table()
for entry in content:
    if entry["mx"] == 10 and entry["method"] == "linear":
        with open( db.outfile( entry) ) as f:
            output = yaml.full_load(f)
            print( output)
```

```python
content = db.table()
df = pd.json_normalize(content)
df
```

```python
table=list()
for data in content :
    with open( db.outfile( data)) as f:
        output = yaml.full_load(f)
        table.append( output)
df = pd.json_normalize(table)
df
```

```python
names = ["centered", "divCentered", "directLap", "divCentered_vol", "directLap_vol"]#, "Tp_mass_err", "Tp_TV_err"]
columns = []
for name in names :
    columns.append( "Guenter."+name)
dflist = []
string_mlist = []
string_methodlist = []
string_dflist = []
for method in methods:
    for m in support:
        if m == 1  and method == "linear-nearest":
            continue
        if m == 1  and method == "linear-linear":
            continue
        df_dir = df[["method", "mx", "Nx", "Nz"]+columns].copy()
        df_dir = df_dir.loc [ df_dir["mx"]==m]
        df_dir = df_dir.loc[ df_dir["method"] == method]
        df_dir = df_dir.sort_values( by=['mx','Nx','Nz']) # sort
        for name in names:
            df_dir.insert(df_dir.columns.get_loc("Guenter."+name)+1,column="order."+name,
                          value=df_dir.loc[:,"Guenter."+name])
            df_dir["order."+name] = (np.log ( np.abs(df_dir["order."+name]/ df_dir["order."+name].shift(1)))/
                np.log( df_dir["Nz"].shift(1)/df_dir["Nz"]))
        df_dir.drop(['method','mx'],axis=1,inplace=True)
        df_dir.rename( columns={"Nx": r"$N$", "Nz": r"$N_\varphi$"},inplace=True)
        df_dir.set_index([r'$N$',r'$N_\varphi$'], inplace=True)
        headers = [ "$\nabla_\parallel^{CC}$ Eq.~\eqref{eq:centered}",
                    "$\nabla \cdot ( {\mathbf{\hat b}} f)^{CC}$ Eq.~\eqref{eq:divCentered}",
                    "$\Delta_\parallel^{CC}$ Eq.~\eqref{eq:laplace}",
                    "$\int dV \nabla \cdot ( {\mathbf{\hat b}} f)^{CC} $",
                    "$\int dV \Delta_\parallel^{CC} f $",
                    #"$\int dV F^+$",
                    #"$\int dV |\nabla_\perp F^+|$"
                  ]
        df_dir.columns=pd.MultiIndex.from_product([headers,["error", "order"] ])
        dflist.append(df_dir)
        string_df_dir = df_dir.copy()
        for header in headers :  
            string_df_dir[header, "error"]=string_df_dir[header, "error"].apply( errorToString)
            string_df_dir[header, "order"]=string_df_dir[header, "order"].apply( orderToString)
        string_dflist.append(string_df_dir)
        string_methodlist.append( method)
        string_mlist.append(str(m))
```

```python
for i in range(0,len(string_dflist)):
    display( string_dflist[i])
    print( "Method: ", string_methodlist[i])
    print( "mx:     ", string_mlist[i])
    filename='ds_guenter_'+string_methodlist[i]+string_mlist[i]+'.tex'
    with open(filename, 'wb') as f:
        f.write(bytes(string_dflist[i].to_latex(
            escape=False,
            #column_format='llp{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}'
            column_format='llllllllllll',
                                   bold_rows=True),'UTF-8'))
```
