Here, we generate the table that can be displayed in parallel.pdf

```python
import json
import simplesimdb as database
import yaml
import numpy as np
import pandas as pd
from IPython.display import display, HTML # to display tables in for loop

pd.set_option('display.float_format', lambda x: '%.2e' % x)
```

```python
#define conversion function 
def orderToString(x): 
    if np.isnan(x) : return 'n/a'
    return'%.2f'% x

def errorToString(x):
    return '%.2e' % x
```

```python
# Create Database manger
db = database.Manager( directory = 'data_ds_t',
        filetype='yaml', executable='./execute.sh')
```

```python
# First generate some data
support=[1,10]
methods=["dg","linear", "cubic"]
resolution = [ (10,5), (16,10), (26,20), (40,40), (64,80) ]#, (100,160)]

# create simulation database
for method in methods :
    for m in support :
        for n in resolution :
            inputfile = { "n": 3, "Nx" : n[0], "Ny" : n[0], "Nz" : n[1],
                "mx" : m, "my" : m, "method" : method }
            print( inputfile)
            db.create( inputfile, error="display")
```

```python
content = db.table()
for entry in content:
    if entry["mx"] == 10 and entry["method"] == "linear":
        with open( db.outfile( entry) ) as f:
            output = yaml.full_load(f)
            print( output["Dirichlet"])
```

```python
content = db.table()
df = pd.io.json.json_normalize(content)
df
```

```python
table=list()
for data in content :
    with open( db.outfile( data)) as f:
        output = yaml.full_load(f)
        table.append( output)
df = pd.io.json.json_normalize(table)
df
```

Now we compute the order of the schemes for given boundary condition and m and change the header to something nice

```python
names = ["centered", "divCentered", "directLap", "invCenteredLap"]
#names = ["centered_bc_along", "dss_bc_along", "divCentered", "invCenteredLap"]

dflist = []
string_dflist = []
string_mlist = []
string_methodlist = []
string_bclist = []
for bc in ["Dirichlet", "Neumann"]:
    columns = []
    for name in names :
        columns.append( bc+"."+name)
    for method in methods:
        for m in support:
            df_dir = df[["method", "mx", "Nx", "Nz"]+columns]
            df_dir = df_dir.loc [ df_dir["mx"]==m]
            df_dir = df_dir.loc[ df_dir["method"] == method]
            df_dir = df_dir.sort_values( by=['mx','Nx','Nz']) # sort
            for name in names:
                df_dir.insert(df_dir.columns.get_loc(bc+"."+name)+1,column="order."+name, value=df_dir.loc[:,bc+"."+name])
                df_dir["order."+name] = (np.log ( df_dir["order."+name]/ df_dir["order."+name].shift(1))/
                    np.log( df_dir["Nz"].shift(1)/df_dir["Nz"]))
            df_dir.drop(['method','mx'],axis=1,inplace=True)
            df_dir.set_index(['Nx','Nz'], inplace=True)
            headers = [ "$\nabla_\parallel$ Eq.~\eqref{eq:paralleldis}",
                        "$\nabla \cdot ( {\mathbf{\hat b}} f)$",
                        "$\Delta_\parallel$ Eq.~\eqref{eq:second_order}",
                        "$(1-\Delta_\parallel)^{-1} f$"]
            df_dir.columns=pd.MultiIndex.from_product([headers,["error", "order"] ])
            dflist.append(df_dir)
            #We would like the order to be displayed differently.
            #In pandas we must convert the columns to strings to do that.
            string_df_dir = df_dir.copy()
            for header in headers :
                string_df_dir[header, "error"]=string_df_dir[header, "error"].apply( errorToString)
                string_df_dir[header, "order"]=string_df_dir[header, "order"].apply( orderToString)
            string_dflist.append(string_df_dir)
            string_methodlist.append( method)
            string_mlist.append(str(m))
            string_bclist.append(bc)
```

```python
for i in range(0,len(string_dflist)):
    display( string_dflist[i])
    print( "Method: ", string_methodlist[i])
    print( "mx:     ", string_mlist[i])
    print( "BC:     ", string_bclist[i])
    filename='ds_'+string_methodlist[i]+string_mlist[i]+string_bclist[i]+'.tex'
    with open(filename, 'wb') as f:
        f.write(bytes(string_dflist[i].to_latex(
            escape=False,column_format='llp{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}p{1.5cm}p{1.2cm}',
                                   bold_rows=True),'UTF-8'))
```
